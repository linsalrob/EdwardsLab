```python
import os
import sys

import re
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import PatchCollection
import pandas as pd
import seaborn as sns
import json
```

```python
fig, ax = plt.subplots(figsize=(18, 12), nrows=2, ncols=2, sharey=True)


for i, j in enumerate(["EagleRay", "PortJackson", "TigerSharks", "NorfolkWater"]):
    df = pd.read_csv(f"Emma_subsystems/{j}_level1_norm_all.tsv", delimiter="\t")
    df = df.set_index("Unnamed: 0")
    axx = 0
    axy = i
    if i > 1:
        axx=1
        axy-=2
    
    sns.heatmap(data=df, ax=ax[axx, axy], cmap="rocket_r")
    ax[axx, axy].set_title(j)
    ax[axx, axy].set_ylabel("")
plt.tight_layout()
fig.savefig("subsystems.png")
    
```
