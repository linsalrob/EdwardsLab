# Plot some histograms of data

```python
import os
import sys

import re
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import PatchCollection
import pandas as pd
import seaborn as sns
import json
```

```python
with open("Lactobacillus_prophage_lengths2.json", 'r') as jf:
    d = json.load(jf)
```

```python
lens = []
for w in d:
    lens.append(len(d[w]))
print(sorted(lens))
```

```python
# this should be doable as a filter, but I can't figure out how
data = {}
for w in d:
    if (len(d[w]) > 500):
        data[w] = list(filter(lambda x: x > 0 and x < 80000, d[w]))
data.keys()
```

```python
c = 0
for w in data:
    print(f"{w}\t{len(data[w])}")
    c+=len(data[w])
print(f"Total: {c}")
```

```python
c = 0
for w in data:
    print(f"{w}\t{len(data[w])}")
    c+=len(data[w])
print(f"Total: {c}")
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
for w in data:
    sns.kdeplot(data[w], shade=True, ax=ax, label=w)
g = ax.legend()
g = ax.set_xlabel("Genome Size (bp)")
ax.set_xlim(0,80000)
fig.savefig("lactobacillus_prophages.png")
```

```python
with open("Bacteroides_prophage_lengths.json", 'r') as jf:
    d = json.load(jf)

lens = []
for w in d:
    lens.append(len(d[w]))
print(sorted(lens))
```

```python
# this should be doable as a filter, but I can't figure out how
data = {}
for w in d:
    if w == 'other':
        continue
    if (len(d[w]) > 1000):
        data[w] = list(filter(lambda x: x > 0 and x < 80000, d[w]))
data.keys()
```

```python
c = 0
for w in data:
    print(f"{w}\t{len(data[w])}")
    c+=len(data[w])
print(f"Total: {c}")
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
for w in data:
    sns.kdeplot(data[w], shade=True, ax=ax, label=w)
g = ax.legend()
g = ax.set_xlabel("Genome Size (bp)")
ax.set_xlim(0,80000)
plt.tight
fig.savefig("bacteroides_prophages.png")
```
